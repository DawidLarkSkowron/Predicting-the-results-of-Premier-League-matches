```python
folder='Datasets/'
data_1 = pd.read_csv(folder +'2010-11.csv')
data_2 = pd.read_csv(folder +'2011-12.csv')
data_3 = pd.read_csv(folder +'2012-13.csv')
data_4 = pd.read_csv(folder +'2013-14.csv')
data_5 = pd.read_csv(folder +'2014-15.csv')
data_6 = pd.read_csv(folder +'2015-16.csv')
data_7 = pd.read_csv(folder +'2016-17.csv')
data_8 = pd.read_csv(folder +'2017-18.csv')
data_9 = pd.read_csv(folder +'2018-19.csv')
data_10 = pd.read_csv(folder +'2019-20.csv')
data_11 = pd.read_csv(folder +'2020-21.csv')
data_12 = pd.read_csv(folder +'2021-22.csv')
```

```python
columns_req = ['Date','HomeTeam','AwayTeam','FTHG','FTAG','FTR']
playing_statistics_1 = data_1[columns_req]                      
playing_statistics_2 = data_2[columns_req]
playing_statistics_3 = data_3[columns_req]
playing_statistics_4 = data_4[columns_req]
playing_statistics_5 = data_5[columns_req]
playing_statistics_6 = data_6[columns_req]
playing_statistics_7 = data_7[columns_req]
playing_statistics_8 = data_8[columns_req]
playing_statistics_9 = data_9[columns_req]
playing_statistics_10 = data_10[columns_req]
playing_statistics_11 = data_11[columns_req]   
playing_statistics_12 = data_12[columns_req]
```

```python
def Calculate_Scored_Goals(playing_stat):
    # Zbierz unikalne drużyny
    teams = playing_stat['HomeTeam'].unique()
    # Przygotuj słownik na gole zdobyte przez każdą drużynę
    goals_dict = {team: [] for team in teams}

    # Iteruj po wierszach DataFrame i przypisuj gole do odpowiednich drużyn
    for _, row in playing_stat.iterrows():
        goals_dict[row['HomeTeam']].append(row['FTHG'])
        goals_dict[row['AwayTeam']].append(row['FTAG'])

    # Zamień słownik na DataFrame: drużyny jako wiersze, kolejki jako kolumny
    df_goals = pd.DataFrame.from_dict(goals_dict, orient='index')
    df_goals.columns = range(1, df_goals.shape[1] + 1)
    df_goals.insert(0, 0, 0)  # Dodaj kolumnę zerową

    # Kumuluj gole zdobyte do danej kolejki
    df_goals = df_goals.cumsum(axis=1)
    return df_goals

def Calculate_Conceded_Goals(playing_stat):
    # Zbierz unikalne drużyny
    teams = playing_stat['HomeTeam'].unique()
    # Przygotuj słownik na stracone gole
    conceded_dict = {team: [] for team in teams}

    # Iteruj po wierszach i przypisuj stracone gole
    for _, row in playing_stat.iterrows():
        conceded_dict[row['HomeTeam']].append(row['FTAG'])
        conceded_dict[row['AwayTeam']].append(row['FTHG'])

    # Zamień słownik na DataFrame: drużyny jako wiersze, kolejki jako kolumny
    df_conceded = pd.DataFrame.from_dict(conceded_dict, orient='index')
    df_conceded.columns = range(1, df_conceded.shape[1] + 1)
    df_conceded.insert(0, 0, 0)  # Dodaj kolumnę zerową

    # Kumulacja straconych goli po kolejkach
    df_conceded = df_conceded.cumsum(axis=1)
    return df_conceded

def Cumulative_Goals_Stats(playing_stat):
    # Pobierz DataFrame z bramkami straconymi i zdobytymi
    goals_conceded = Calculate_Conceded_Goals(playing_stat)
    goals_scored = Calculate_Scored_Goals(playing_stat)

    home_scored, away_scored = [], []
    home_conceded, away_conceded = [], []
    matchweek = 0

    # Przechodzimy przez każdy mecz i pobieramy wartości z odpowiednich DataFrame
    for idx, row in playing_stat.iterrows():
        home = row['HomeTeam']
        away = row['AwayTeam']
        home_scored.append(goals_scored.loc[home, matchweek])
        away_scored.append(goals_scored.loc[away, matchweek])
        home_conceded.append(goals_conceded.loc[home, matchweek])
        away_conceded.append(goals_conceded.loc[away, matchweek])

        if (idx + 1) % 10 == 0:
            matchweek += 1

    playing_stat['HTGS'] = home_scored
    playing_stat['ATGS'] = away_scored
    playing_stat['HTGC'] = home_conceded
    playing_stat['ATGC'] = away_conceded

    return playing_stat


playing_statistics_1 = Cumulative_Goals_Stats(playing_statistics_1)
playing_statistics_2 = Cumulative_Goals_Stats(playing_statistics_2)
playing_statistics_3 = Cumulative_Goals_Stats(playing_statistics_3)
playing_statistics_4 = Cumulative_Goals_Stats(playing_statistics_4)
playing_statistics_5 = Cumulative_Goals_Stats(playing_statistics_5)
playing_statistics_6 = Cumulative_Goals_Stats(playing_statistics_6)
playing_statistics_7 = Cumulative_Goals_Stats(playing_statistics_7)
playing_statistics_8 = Cumulative_Goals_Stats(playing_statistics_8)
playing_statistics_9 = Cumulative_Goals_Stats(playing_statistics_9)
playing_statistics_10 = Cumulative_Goals_Stats(playing_statistics_10)
playing_statistics_11 = Cumulative_Goals_Stats(playing_statistics_11)
playing_statistics_12 = Cumulative_Goals_Stats(playing_statistics_12)
```

```python
def Calculate_Points(result):
    # Zamiana wyniku meczu na punkty
    points_map = {'W': 3, 'D': 1}
    return points_map.get(result, 0)

def Get_Cumulative_Points(matchres):
    # Zamiana wyników meczowych na punkty
    points_df = matchres.applymap(Calculate_Points)
    # Dodanie kolumny zerowej z początkowymi punktami
    points_df.insert(0, 0, 0)
    # Kumulacja punktów po kolejkach
    points_df = points_df.cumsum(axis=1)
    return points_df

def Calculate_Teams_Form(playing_stat):
    # Zbierz unikalne drużyny
    teams = playing_stat['HomeTeam'].unique()
    results = {team: [] for team in teams}

    # Iteruj po wierszach DataFrame
    for _, row in playing_stat.iterrows():
        if row['FTR'] == 'H':
            results[row['HomeTeam']].append('W')
            results[row['AwayTeam']].append('L')
        elif row['FTR'] == 'A':
            results[row['AwayTeam']].append('W')
            results[row['HomeTeam']].append('L')
        else:
            results[row['HomeTeam']].append('D')
            results[row['AwayTeam']].append('D')

    # Zamień słownik na DataFrame: drużyny jako wiersze, kolejki jako kolumny
    df_results = pd.DataFrame.from_dict(results, orient='index')
    df_results.columns = range(1, df_results.shape[1] + 1)
    return df_results

def Get_All_Points(playing_stat):
    # Oblicz wyniki meczów i skumulowane punkty
    match_results = Calculate_Teams_Form(playing_stat)
    cumulative_points = Get_Cumulative_Points(match_results)

    home_points, away_points = [], []
    matchweek = 0

    # Iteruj po każdym meczu i przypisz punkty gospodarzom i gościom
    for idx, row in playing_stat.iterrows():
        home = row['HomeTeam']
        away = row['AwayTeam']
        home_points.append(cumulative_points.loc[home, matchweek])
        away_points.append(cumulative_points.loc[away, matchweek])
        if (idx + 1) % 10 == 0:
            matchweek += 1

    playing_stat['HTP'] = home_points
    playing_stat['ATP'] = away_points
    return playing_stat

playing_statistics_1 = Get_All_Points(playing_statistics_1)
playing_statistics_2 = Get_All_Points(playing_statistics_2)
playing_statistics_3 = Get_All_Points(playing_statistics_3)
playing_statistics_4 = Get_All_Points(playing_statistics_4)
playing_statistics_5 = Get_All_Points(playing_statistics_5)
playing_statistics_6 = Get_All_Points(playing_statistics_6)
playing_statistics_7 = Get_All_Points(playing_statistics_7)
playing_statistics_8 = Get_All_Points(playing_statistics_8)
playing_statistics_9 = Get_All_Points(playing_statistics_9)
playing_statistics_10 = Get_All_Points(playing_statistics_10)
playing_statistics_11 = Get_All_Points(playing_statistics_11)
playing_statistics_12 = Get_All_Points(playing_statistics_12)
```

```python
def Get_WLD_Form(playing_stat, num):
    match_results = Calculate_Teams_Form(playing_stat)
    form_final = match_results.copy()
    # Użyj poprawnego zakresu kolumn
    for col in range(num, match_results.shape[1]):
        form_final[col] = [
            ''.join([match_results.iloc[row_idx, col - j] for j in range(num)])
            for row_idx in range(match_results.shape[0])
        ]
    return form_final

def Add_MW_Form(playing_stat, num):
    # Pobierz formę drużyn (ciągi wyników) dla zadanej liczby meczów
    form = Get_WLD_Form(playing_stat, num)
    num_matches = num * 10
    total_matches = len(playing_stat)

    # Listy na formę gospodarzy i gości, domyślnie 'M' dla braku danych
    home_form = ['X'] * num_matches
    away_form = ['X'] * num_matches

    matchweek = num
    for idx in range(num_matches, total_matches):
        home_team = playing_stat.iloc[idx]['HomeTeam']
        away_team = playing_stat.iloc[idx]['AwayTeam']

        # Dodaj wynik ostatniego meczu z formy gospodarzy i gości
        home_form.append(form.at[home_team, matchweek][num - 1])
        away_form.append(form.at[away_team, matchweek][num - 1])

        # Po każdych 10 meczach przechodzimy do kolejnej kolejki
        if (idx + 1) % 10 == 0:
            matchweek += 1

    # Dodaj kolumny do DataFrame
    playing_stat[f'HM{num}'] = home_form
    playing_stat[f'AM{num}'] = away_form

    return playing_stat

def Add_MW_Form_Df(playing_statistics):
    # Dodaj historię wyników dla num od 1 do 5 w pętli
    for num in range(1, 6):
        playing_statistics = Add_MW_Form(playing_statistics, num)
    return playing_statistics

playing_statistics_1 = Add_MW_Form_Df(playing_statistics_1)
playing_statistics_2 = Add_MW_Form_Df(playing_statistics_2)
playing_statistics_3 = Add_MW_Form_Df(playing_statistics_3)
playing_statistics_4 = Add_MW_Form_Df(playing_statistics_4)
playing_statistics_5 = Add_MW_Form_Df(playing_statistics_5)
playing_statistics_6 = Add_MW_Form_Df(playing_statistics_6)
playing_statistics_7 = Add_MW_Form_Df(playing_statistics_7)
playing_statistics_8 = Add_MW_Form_Df(playing_statistics_8)
playing_statistics_9 = Add_MW_Form_Df(playing_statistics_9)
playing_statistics_10 = Add_MW_Form_Df(playing_statistics_10)
playing_statistics_11 = Add_MW_Form_Df(playing_statistics_11)
playing_statistics_12 = Add_MW_Form_Df(playing_statistics_12)
```

```python
# Przestawianie kolumn

cols = ['Date', 'HomeTeam', 'AwayTeam', 'FTHG', 'FTAG', 'FTR', 'HTGS', 'ATGS', 'HTGC', 'ATGC', 'HTP', 'ATP', 'HM1', 'HM2', 'HM3',
        'HM4', 'HM5', 'AM1', 'AM2', 'AM3', 'AM4', 'AM5' ]

playing_statistics_1 = playing_statistics_1[cols]
playing_statistics_2 = playing_statistics_2[cols]
playing_statistics_3 = playing_statistics_3[cols]
playing_statistics_4 = playing_statistics_4[cols]
playing_statistics_5 = playing_statistics_5[cols]
playing_statistics_6 = playing_statistics_6[cols]
playing_statistics_7 = playing_statistics_7[cols]
playing_statistics_8 = playing_statistics_8[cols]
playing_statistics_9 = playing_statistics_9[cols]
playing_statistics_10 = playing_statistics_10[cols]
playing_statistics_11 = playing_statistics_11[cols]
playing_statistics_12 = playing_statistics_12[cols]
```

```python
# Łączenie wszystkich DataFrame w jeden, korzystając z listy i rozpakowania
all_stats = [
    playing_statistics_1, playing_statistics_2, playing_statistics_3, playing_statistics_4,
    playing_statistics_5, playing_statistics_6, playing_statistics_7, playing_statistics_8,
    playing_statistics_9, playing_statistics_10, playing_statistics_11, playing_statistics_12
]

playing_stat = pd.concat(all_stats, ignore_index=True)

def Get_MW(playing_stat):
    num_matches = len(playing_stat)
    matches_per_week = 10
    matchweeks = [(i // matches_per_week) + 1 for i in range(num_matches)]
    playing_stat['MW'] = matchweeks
    return playing_stat

playing_stat = Get_MW(playing_stat)

playing_statistics_1 = Get_MW(playing_statistics_1)
playing_statistics_2 = Get_MW(playing_statistics_2)
playing_statistics_3 = Get_MW(playing_statistics_3)
playing_statistics_4 = Get_MW(playing_statistics_4)
playing_statistics_5 = Get_MW(playing_statistics_5)
playing_statistics_6 = Get_MW(playing_statistics_6)
playing_statistics_7 = Get_MW(playing_statistics_7)
playing_statistics_8 = Get_MW(playing_statistics_8)
playing_statistics_9 = Get_MW(playing_statistics_9)
playing_statistics_10 = Get_MW(playing_statistics_10)
playing_statistics_11 = Get_MW(playing_statistics_11)
playing_statistics_12 = Get_MW(playing_statistics_12)


def Get_Points_Form(string):
    # Zwraca sumę punktów za podany ciąg wyników meczów (W, D, L)
    return sum(Calculate_Points(letter) for letter in string)

playing_stat['HTFormPtsStr'] = playing_stat[[f'HM{i}' for i in range(1, 6)]].agg(''.join, axis=1)
playing_stat['ATFormPtsStr'] = playing_stat[[f'AM{i}' for i in range(1, 6)]].agg(''.join, axis=1)

playing_stat['HTFormPts'] = playing_stat['HTFormPtsStr'].map(Get_Points_Form)
playing_stat['ATFormPts'] = playing_stat['ATFormPtsStr'].map(Get_Points_Form)


def Get_3game_WinStreak(string):
    # Zwraca 1 jeśli ostatnie 3 mecze to zwycięstwa, w przeciwnym razie 0
    return int(string[-3:] == 'WWW')

def Get_5game_WinStreak(string):
    # Zwraca 1 jeśli ostatnie 5 meczów to zwycięstwa, w przeciwnym razie 0
    return int(string[-5:] == 'WWWWW')

def Get_3game_LoseStreak(string):
    # Zwraca 1 jeśli ostatnie 3 mecze to porażki, w przeciwnym razie 0
    return int(string[-3:] == 'LLL')

def Get_5game_LoseStreak(string):
    # Zwraca 1 jeśli ostatnie 5 meczów to porażki, w przeciwnym razie 0
    return int(string[-5:] == 'LLLLL')


for prefix in ['HT', 'AT']:
    playing_stat[f'{prefix}WinStreak3'] = playing_stat[f'{prefix}FormPtsStr'].map(Get_3game_WinStreak)
    playing_stat[f'{prefix}WinStreak5'] = playing_stat[f'{prefix}FormPtsStr'].map(Get_5game_WinStreak)
    playing_stat[f'{prefix}LossStreak3'] = playing_stat[f'{prefix}FormPtsStr'].map(Get_3game_LoseStreak)
    playing_stat[f'{prefix}LossStreak5'] = playing_stat[f'{prefix}FormPtsStr'].map(Get_5game_LoseStreak)

playing_stat.keys()
```

```python
# Wyliczanie różnicy bramek i punktów w jednej operacji
playing_stat = playing_stat.assign(
    HTGD = playing_stat['HTGS'] - playing_stat['HTGC'],
    ATGD = playing_stat['ATGS'] - playing_stat['ATGC'],
    DiffPts = playing_stat['HTP'] - playing_stat['ATP'],
    DiffFormPts = playing_stat['HTFormPts'] - playing_stat['ATFormPts']
)
```

```python
# Skalowanie wybranych kolumn przez numer kolejki (MW)
cols = ['HTGD', 'ATGD', 'DiffPts', 'DiffFormPts', 'HTP', 'ATP']
playing_stat['MW'] = playing_stat['MW'].astype(float)

playing_stat[cols] = playing_stat[cols].div(playing_stat['MW'], axis=0)
```
