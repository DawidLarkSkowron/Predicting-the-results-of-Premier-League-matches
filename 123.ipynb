```python
import matplotlib.pyplot as plt
import numpy as np

# Dane do wykresu
klasy = ['H', 'D', 'A']
przed_balansowaniem = [2044, 1107, 1409]
po_smote = [2044, 2044, 2044]
po_undersamplingu = [1107, 1107, 1107]

x = np.arange(len(klasy))
szerokosc = 0.25  # NIE większa! 3*szerokosc < 1

fig, ax = plt.subplots(figsize=(10, 7))

ax.bar(x - szerokosc, przed_balansowaniem, szerokosc, label='Przed balansowaniem', color='gray')
ax.bar(x, po_smote, szerokosc, label='Po SMOTE', color='blue')
ax.bar(x + szerokosc, po_undersamplingu, szerokosc, label='Po undersamplingu', color='orange')

ax.set_xlabel('Klasa wyniku (FTR)')
ax.set_ylabel('Liczba próbek')
ax.set_xticks(x)
ax.set_xticklabels(klasy)

# Najpierw legenda, potem tytuł!
leg = ax.legend(
    loc='lower center',
    bbox_to_anchor=(0.5, 1.02),
    ncol=3,
    frameon=True,
    fontsize=12
)

# Teraz tytuł, żeby nie nachodził na legendę
plt.title('Porównanie liczebności klas przed i po balansowaniu', pad=45, fontsize=15)

plt.tight_layout(rect=[0, 0, 1, 0.93])  # zostawia miejsce na legendę i tytuł
plt.show()
```
